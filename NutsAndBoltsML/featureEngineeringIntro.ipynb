```python
# Fetch the data

churn_data_raw = pd.read_csv('./raw_data/Churn_Modelling.csv')
churn_data_raw.head(20)
```

```python
churn_data = churn_data_raw.drop(columns=['RowNumber','CustomerId','Surname','Gender'])
churn_data.head()
```

create a Loyalty feature that represents the percentage of each customer's life that they were customers.

```python
churn_data['Loyalty'] = round(churn_data['Tenure'] / churn_data['Age'],4)
churn_data.head()
```

```python
churn_data.sort_values(by= 'Loyalty',ascending=False)
```
