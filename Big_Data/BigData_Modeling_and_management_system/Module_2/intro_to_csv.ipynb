## Introduction to CSV

```python
import pandas as pd
import os 

parent_wd = os.path.abspath(os.path.join(os.getcwd(),'..'))
folder_name = 'Raw_data'
file_name = 'census.csv'
file_path = os.path.join(parent_wd,folder_name,file_name)

first_csv = pd.read_csv(file_path, encoding='latin1')
```

```python
first_csv.columns
```

```python
census_by_state = first_csv.groupby(['STNAME']).agg({'CENSUS2010POP':'sum',''}).reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(1,1,figsize=(17,8))
sns.barplot(
    data = census_by_state,
    x = 'STNAME',
    y = 'CENSUS2010POP'
)

plt.xticks(rotation = 45, horizontalalignment = 'right')
```
