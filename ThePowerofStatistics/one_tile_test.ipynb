# One tail test

#### Example

 The company claims that at least 80% of its customers are satisfied with their shopping experience. You survey a random sample of 100 customers. According to the survey, 73% of customers say they are satisfied. Based on the survey data, you conduct a z-test to evaluate the claim that at least 80% of customers are satisfied. 

```python
from scipy import stats
import math
```

```python
## This is a case when the population proportion aproach works
## the example has a percentace of individuals or elements share a characteristic

po = 0.8
p = 0.73
sample_size = 100


# H0 >= 0.8
# Ha < 0.8

standard_error = (( po * (1 - po)  ) / sample_size) ** 0.5
print(f"{standard_error:.2f}")

z_score = (p - po) / standard_error

print(f"{z_score:.2f}")
```

```python
p_value = stats.norm.cdf(z_score)

print(f"{p_value:.5f}")
```

```python
significance_level = 0.05 
print(f"The p_value for the problem is {p_value:.5f} and the significace level {significance_level:.4f}")
```
