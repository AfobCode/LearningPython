```python
from scipy.stats import binom 
```

```python
# Based on the problem 
# p = 10% 
# n = 3 customers

p_return = 0.1
n_trials = 3
```

```python
# Now use binom funtion to get the probability 
# of 0 returns 

n_success= 0

prob_zero = binom.pmf(n_success,n_trials,p_return)
print(prob_zero)
```

```python
# Now check the probability of one return 

k_success = 1

prob_one = binom.pmf(k_success,n_trials,p_return)
print(prob_one)
```

```python
# Now check the probability of TWO returns

k_success = 2

prob_two = binom.pmf(k_success,n_trials,p_return)
print(prob_two)
```

```python
# Now check the probability of Three returns

k_success = 3

prob_three = binom.pmf(k_success,n_trials,p_return)
print(prob_three)
```
