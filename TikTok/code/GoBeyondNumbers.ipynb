## Task 4. Determine outliers

```python
columns_check = ["video_view_count", "video_like_count","video_share_count","video_download_count","video_comment_count"]

for column in columns_check:
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    median_col = data[column].median()
    
    upper_limit = median_col + 1.5*iqr
    
    outliers = data[data[column] > upper_limit]
    outliers_count = outliers[column].count()
    
    print(f"Number of outliers, {column}: {outliers_count}")
```

```python
# Create a scatterplot of `video_view_count` versus `video_like_count` according to 'claim_status'

plt.figure(figsize = (10,6))

sns.scatterplot(
    data = data,
    x = "video_view_count",
    y = "video_like_count",
    hue = "claim_status",
    s = 5, # Dot size 
    alpha = 0.3
)
```
