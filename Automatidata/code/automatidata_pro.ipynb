### Task 2c. Understand the data - Investigate the variables

```python
# Sort your first variable (trip_distance) from maximum to minimum value, do the values seem normal?

raw_data.sort_values(by="trip_distance",ascending=False)[["id","trip_distance","fare_amount","total_amount"]].head(10)
```

```python
# Sort the data by total amount and print the top 20 values
raw_data.sort_values(by="total_amount",ascending=False)[["id","trip_distance","fare_amount","total_amount"]].head(15)
```

```python
# Sort the data by total amount and print the bottom 20 values
raw_data.sort_values(by="total_amount", ascending=False)[["id","trip_distance","payment_type","fare_amount","total_amount"]].tail(20)
```

```python
# How many of each payment type are represented in the data?
raw_data.value_counts(["payment_type"])
```

```python
# What is the average tip for trips paid for with credit card?
cc_trips = raw_data[raw_data["payment_type"] == 1]
avg_cc_payment = round(cc_trips["tip_amount"].mean(),4)
print(avg_cc_payment)

# What is the average tip for trips paid for with cash?
cash_trips = raw_data[raw_data["payment_type"] == 2]
avg_cash_payment = round(cash_trips["tip_amount"].mean(),4)
sum_cash_payment = round(cash_trips["tip_amount"].sum(),4)
print(avg_cash_payment,sum_cash_payment)
cash_trips
```

```python
# How many times is each vendor ID represented in the data?
raw_data.value_counts("VendorID")
```

```python
# What is the mean total amount for each vendor?
raw_data.groupby(["VendorID"])["total_amount"].mean()
```

```python
# Filter the data for credit card payments only
cc_mask = raw_data["payment_type"] == 1
cc_trips = raw_data[cc_mask]
cc_trips
```

```python
# Calculate the average tip amount for each passenger count (credit card payments only)

cc_tip_per_passengers = cc_trips.groupby("passenger_count")["tip_amount"].mean()
cc_tip_per_passengers = round(cc_tip_per_passengers,3)
print(cc_tip_per_passengers)
```
