## Boolean masking

a filtering technique that overlays a boolean grid onto a dataframe in order to select only the values in the dataframe that align with the True values of the Grid

```python
# Instantiate a dictionary of planetary data.
data = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars',
                   'Jupiter', 'Saturn', 'Uranus', 'Neptune'],
       'radius_km': [2440, 6052, 6371, 3390, 69911, 58232,
                     25362, 24622],
       'moons': [0, 0, 1, 2, 80, 83, 27, 14]
        }

planets = pd.DataFrame(data)
print(planets)
```

```python
mask = planets["moons"] > 20
print(mask)
print("="*35)

planets_20_moons = planets[mask]
print(planets_20_moons)
```

```python
pla_20m = planets[planets["moons"] > 20]
print(pla_20m)
print("="*35)

print(planets)
```

```python
# It can also use the logic to filter by conditioners like or "|"
# the and "&", and the nefation mark "~" if the mask will be built 
# in a different statement should be write into parenthesis ()

mask = (planets["moons"] > 50) | (planets["moons"] < 3)
# This mask will return a True if the row has a moon value greather than 50 OR lower than 3

pl_3m_50m = planets[mask]
print(pl_3m_50m)
print("="*35)
print(planets)
```

```python
# And Mask

mask = (planets["moons"] > 10) & (planets["radius_km"] > 25000)
print(planets[mask])

print("="*45)
mask_not = (planets["moons"] < 15) & ~(planets["radius_km"] > 25000)
print(planets[mask_not])
```

## Use groupby()
A pandas DataFrame method that groups rows of the dataframe together based on their values at one or more columns, which allows further analysis of the groups

```python
# Instantiate a dictionary of planetary data.
data = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars',
                   'Jupiter', 'Saturn', 'Uranus', 'Neptune'],
        'radius_km': [2440, 6052, 6371, 3390, 69911, 58232,
                     25362, 24622],
        'moons': [0, 0, 1, 2, 80, 83, 27, 14],
        'type': ['terrestrial', 'terrestrial', 'terrestrial', 'terrestrial',
                 'gas giant', 'gas giant', 'ice giant', 'ice giant'],
        'rings': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes','yes'],
        'mean_temp_c': [167, 464, 15, -65, -110, -140, -195, -200],
        'magnetic_field': ['yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes']
        }

planets = pd.DataFrame(data)
print(planets)
```

```python
planets_by_type = planets.groupby(["type"]).sum(numeric_only=True)
print(planets_by_type)
```

```python
planets_by_type_moon = planets_by_type.loc[:,['moons']]
print(planets_by_type_moon)
print("*"*18)
planets_by_type_moon_2 = planets_by_type_moon[['moons']]
print(planets_by_type_moon_2)
```

```python
print(planets)
planets.groupby(["type","magnetic_field"]).mean(numeric_only=True)
```

```python
print(planets)
planets.groupby(["rings","type"]).mean(numeric_only=True)
```

## The aggreate function in Pandas
.agg()
Short of the aggregate. A pandas groupby method that allows me to apply multiple calculations to groups of data

Calling this function will allow to pass a list of calculations that I would like to see per column

```python
planets.groupby(['type']).agg({"radius_km":["count","sum","mean"],"moons":["sum","mean"]})
```

## Merging and Joining Data

### Main functions
* concat()
* merge()

```python
import pandas as pd

data_1 = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars'],
        'radius_km': [2440, 6052, 6371, 3390],
        'moons': [0, 0, 1, 2],
        }

data_2 = {'planet': ['Jupiter', 'Saturn', 'Uranus', 'Neptune'],
        'radius_km': [69911, 58232, 25362, 24622],
        'moons': [80, 83, 27, 14],
        }

data_3 = {'planet': ['Earth', 'Mars','Jupiter', 'Saturn', 'Uranus',
                   'Neptune', 'Janssen', 'Tadmor'],
        'type': ['terrestrial', 'terrestrial','gas giant', 'gas giant',
                 'ice giant', 'ice giant', 'super earth','gas giant'],
        'rings': ['no', 'no', 'yes', 'yes', 'yes','yes', 'no', None],
        'mean_temp_c': [15, -65, -110, -140, -195, -200, None, None],
        'magnetic_field': ['yes', 'no', 'yes', 'yes', 'yes', 'yes', None, None],
        'life': [1, 0, 0, 0, 0, 0, 1, 1]
        }

planets_p1 = pd.DataFrame(data_1)
planets_p2 = pd.DataFrame(data_2)
planets_comments = pd.DataFrame(data_3)
print(planets_p1)
print("*"*23)
print(planets_p2)
```

## Concat()

A pandas function that combines data either by adding it horizontally as new columns for existing rows, or vertically as new rows for existing columns

```python
# Concat two df vertically => add rows on the same columns

planets_complete = pd.concat([planets_p1,planets_p2],axis = 0)
planets_complete
```

```python
# Concat two df horizontally => add columns, same amount of rows

planets_compared = pd.concat([planets_p1,planets_p2],axis=1)
planets_compared
```

```python
planets_complete = pd.concat([planets_p1,planets_p2],axis = 0)
print(planets_complete) # row index are the same as the original df for each row
print(".="*15)
print(planets_complete.iloc[[0],:])
print(".="*15)
print(planets_complete.loc[[0],:])
print(".="*15)
#To reset the indexes I can apply the function reset_idex
planets_complete = pd.concat([planets_p1,planets_p2],axis = 0)
planets_complete = planets_complete.reset_index()
planets_complete
print(".="*15)

# and pass the argument drop=True to override the current indexes al leave them new
planets_complete = pd.concat([planets_p1,planets_p2],axis = 0)
planets_complete = planets_complete.reset_index(drop=True)
planets_complete
```
