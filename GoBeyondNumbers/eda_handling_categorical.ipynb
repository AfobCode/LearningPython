```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
```

```python
lighning_info = pd.read_csv("./Raw_data/eda_dataset_cleaning/Files/eda_label_encoding_dataset.csv")
lighning_info.head(10)
```

```python
lighning_info["date"] = pd.to_datetime(lighning_info["date"])

lighning_info["month"] = lighning_info["date"].dt.month
lighning_info["year"] = lighning_info["date"].dt.year
lighning_info["Mth Short"] = lighning_info["date"].dt.month_name().str.slice(stop=3)

lighning_info.head(10)
```

```python
lighning_info_year = lighning_info.groupby(["year"], as_index= False)["number_of_strikes"].sum()
lighning_info_year
```

```python
months_ord = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

lighning_info["month_ctg"] = pd.Categorical(lighning_info["Mth Short"],categories= months_ord, ordered= True)
lighning_info.head(20)
```

```python
lighning_info["Year-str"] = lighning_info["date"].dt.strftime("%Y")
lighning_info.head(10)
```

```python
lighning_info.info()
```

```python
lighning_info_y_by_m = lighning_info.groupby(["year","month_ctg"], as_index=False)["number_of_strikes"].sum()
lighning_info_y_by_m
```

```python
levels = ["Mild", "Scattered", "Heavy", "Severe"]

# Pandas has a method called qcut() which will cut the data into the four quartiles, 25% - 50% - 75% - 100%
# So what I want to do is to categorize the data into a level depending on the quartile range. 
lighning_info_y_by_m["Strike_Level"] = pd.qcut(
    lighning_info_y_by_m["number_of_strikes"], 4, labels = levels
)

lighning_info_y_by_m
```

```python
# Level encode the strike level

lighning_info_y_by_m["Strike_Level_cat"] = lighning_info_y_by_m["Strike_Level"].cat.codes
lighning_info_y_by_m
```

```python
# Get dummy variables (1 or 0) from strike levels

dummy_mtx_lightning =  pd.get_dummies( lighning_info_y_by_m["Strike_Level"])
dummy_mtx_lightning
```

## Pivot Function
Pivot allows me to reshape dataframes into any given index or set of values I want, based on the parameters inside the pivot function

```python
lighning_info_y_by_m.head()
```

```python
lighning_by_mth_plt = lighning_info_y_by_m.pivot(index= "year", columns= "month_ctg", values= "Strike_Level_cat")
lighning_by_mth_plt
```

```python
lighning_by_mth_ctg = lighning_info_y_by_m.pivot(index= "year", columns= "month_ctg", values= "Strike_Level")
lighning_by_mth_ctg
```

```python
ax = sns.heatmap(lighning_by_mth_plt, cmap="Reds")
plt.show()
```

```python
ax = sns.heatmap(lighning_by_mth_plt, cmap="Blues")
colorbar = ax.collections[0].colorbar
colorbar.set_ticks([0,1,2,3])
colorbar.set_ticklabels(levels)
plt.show()
```
